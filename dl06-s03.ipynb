```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x = np.linspace(-2 * np.pi , 2 * np.pi , 10000)
x.shape
```

```python
x = x.reshape(-1 , 1)
```

```python
f_x = np.tanh(x)
f_x
```

```python
f_prime_x = 1 - (f_x ** 2)
f_prime_x
```

```python
plt.plot(x, f_x)
```

```python
plt.plot(x, f_prime_x)
```
